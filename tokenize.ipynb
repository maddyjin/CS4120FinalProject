```python
fname = 'English_philosophy'
corpus = fname.split('_')[0].lower()

with open(f'data/Philosophy/{fname}.txt', 'rb') as f:
    txt = f.read().decode().strip().lower()
```

```python
import glob

lang = 'spanish'
fnames = glob.glob(f'data/*/*{lang}*.txt', recursive=True)

txt = ''

for file in fnames:
    with open(file, 'rb') as f:
        txt += f.read().decode().strip().lower()

txt = txt.replace("\r", "").replace('\n', '')
txt.split()
```

## magic

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
from nltk.corpus import stopwords
import string
stop_words = set(stopwords.words('english'))
```

```python
tokens = word_tokenize(txt, language=corpus)
token_counter = Counter(tokens)
tokens = [word for word in tokens if \
          (word not in stop_words \
           and word not in string.punctuation \
           and token_counter[word] > 2)]
```

```python
print(len(tokens) + 1)
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences=[tokens], vector_size=100, window=5, min_count=1, workers=4)
```

```python
model.wv.most_similar(positive = ['woman', 'king'], negative=['man'])
```
