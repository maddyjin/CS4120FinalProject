```python
fname = 'English_LGBTQ'
corpus = fname.split('_')[0].lower()

with open(f'data/LGBT/{fname}.txt', 'rb') as f:
    txt = f.read().decode().strip().lower()
```

## magic

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
from nltk.corpus import stopwords
import string
from nltk.stem import SnowballStemmer 
stop_words = set(stopwords.words(corpus))
```

```python
tokens = word_tokenize(txt, language=corpus)
token_counter = Counter(tokens)

stemmer = SnowballStemmer(corpus)

tokens = [stemmer.stem(word) for word in tokens if \
          (word not in stop_words \
           and word not in string.punctuation \
           and token_counter[word] > 2)]
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences=[tokens], vector_size=1, window=5, min_count=1, workers=4)
```

```python
model.wv.most_similar(positive = ['woman', 'king'], negative=['man'])
```
